```python
import pandas as pd
import numpy as np
import sklearn
from scipy.stats import stats

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/airquality.csv')

df.head()
```

```python
df.info()
```

```python
# 결측치를 제외한 평균
before_mean = df['Ozone'].mean()

# 결측치를 0으로 대체한 평균
df['Ozone'].fillna(0, inplace=True)
after_mean = df['Ozone'].mean()

print(abs(after_mean - before_mean))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Wind_mm'] = scaler.fit_transform(df[['Wind']])
```

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
df['Wind_st'] = scaler2.fit_transform(df[['Wind']])

mean_mm = df['Wind_mm'].mean()
mean_st = df['Wind_st'].mean()

print(mean_mm, mean_st)
print(abs(mean_mm - mean_st))
```

```python
# 해답
Min = np.min(df['Wind'])
Max = np.max(df['Wind'])
df["min_max"] = round((df['Wind']- Min) / (Max-Min),2)

Mean = np.mean(df['Wind'])
Std = np.std(df['Wind_mm'])
df['z'] = round((df['Wind']- Mean)/ Std, 2)
```

```python
df['min_max'].mean() - df['z'].mean()
```

```python
df.groupby('Month').mean()
```

```python
df.head()
```

```python
df = df.drop(['ID', 'ZIP Code'], axis = 1 )
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('Personal Loan', axis = 1)
y = df['Personal Loan']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import *

preprocessor = Normalizer()

X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
score = []

ran = range(1,25)
for k in ran :
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(X_train, y_train)
  acc = knn.score(X_test, y_test)
  score.append(acc)
```

```python
df['BMI'] = df['Weight'] / (df['Height'] / 100)**2

def cal(BMI):
  if BMI < 18.5:
    return '저체중'
  elif BMI < 23:
    return '정상'
  elif BMI < 25:
    return '과체중'
  elif BMI < 30:
    return '경도비만'
  else :
    return '중등도비만'

df['BMI_range'] = df['BMI'].apply(cal)

print(len(df[df['BMI_range'] == '정상']))
print(len(df[df['BMI_range'] == '과체중']))

```
