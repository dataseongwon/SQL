### 표본의 평균 검정


#### 모집단의 분산(표준편차)을 알고 있는 경우
* Z-test(양측)

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm

data = np.array([23,25,28,30,26,27,29,32,31,28])
```

```python
모평균 = 28
모분산 = 4

유의수준 = 0.05

alternative = "two.sided"

n = len(data)

sample_mean = np.mean(data)

# 검정통계량 계산
test_statistics = (sample_mean - 모평균) / (np.sqrt(모분산) / np.sqrt(n))

# 임계값 계산
critical_value = norm.ppf(1 - 유의수준 / 2)

# p-값 계산
p_value = 2*(1-norm.cdf(abs(test_statistics)))

print(test_statistics)
print(critical_value)
print(p_value)
```

* 분산을 모를 때


```python
alpha = 0.05
alternative = 'two-sided'

# t 검정 수행 , 단일표본검정을 할 수 있는 함수사용
t_statistics, p_value = stats.ttest_1samp(data, popmean = 0, alternative = alternative)

# 95% 신뢰 구간 계산
confidence_interval = stats.t.interval(1-alpha, len(data)-1, loc=np.mean(data),scale = stats.sem(data))

print(t_statistics)
print(p_value)
print(confidence_interval)

```

```python
## 따라하기 예제

import pandas as pd
import numpy as np
from scipy import stats

mtcars = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/mtcars.csv')
```

#### 단일표본 모분산 검정(카이제곱검정)

```python
import numpy as np
from scipy.stats import chi2

data = np.array([14.2, 15.1, 14.8, 15.4, 14.9, 15.2, 14.6, 15.0])

# 가설 모분산은 1과같다. 아니다.

pop_var = 1.0

# 검정통계량 = (n-1) * 표본분산 / 모분산

chi2_statistics = len(data) -1 * np.var(data) / pop_var

degrees_of_freedom = len(data)-1

p_value = 1 - chi2.cdf(chi2_statistics, degrees_of_freedom)

print(p_value)
```

#### 모분산 비에 대한 검정 : 분산이 동일한지 검정


```python
np.random.seed(123)
sample1 = np.random.normal(loc =5, scale=2, size =50)
sample2 = np.random.normal(loc=5, scale=3, size = 50)
```

```python
# 각표본의 분산 계산
var1 = np.var(sample1)
var2 = np.var(sample2)
```

```python
# f검정 통계량 : var1/var2

f = var1 / var2

# f검정실시
p_value = 1-stats.f.cdf(f, len(sample1)-1,len(sample2)-1)

print(p_value)
```

#### 독립성 검정 (카이제곱 검정) : 주로 범주형 데이터에 대한 통계적 가설검정

```python
from scipy.stats import chi2_contingency

data_matrix = np.array([[50, 75, 125, 175], [90, 30, 45, 10]])
```

```python
chi2, p, df, expected = chi2_contingency(data_matrix)

print(chi2, p , df ,expected)
```

```python
from scipy.stats import chi2_contingency


df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/titanic.csv')
```

```python
# 교차표 생성

cross_table = pd.crosstab(df['Sex'], df['Survived'])
```

```python
# 독립성 검정 실시
chi2_contingency(cross_table)
```

```python
def cal(Age):
  if Age < 20:
    return '10대'
  elif Age < 30:
    return '20대'
  elif Age < 40:
    return '30대'
  elif Age < 50:
    return '40대'
  elif Age < 60:
    return '50대'
  else :
    return '60대'
```

```python
# Age가 결측인 경우 평균값으로 대체
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace =True)
```

```python
# 새로운 열 생성
df['Age_c'] = df['Age'].apply(cal)
```

```python
# 교차표 생성
cross_table = pd.crosstab(df['Survived'], df['Age_c'])

# 검정
chi2_contingency(cross_table)

# 귀무가설 기각, 연령대에 따른 생존여부에 차이가 잇음
```
